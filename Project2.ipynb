### How many movies have an average rating over 4.5 overall?

```python
#Group by MovieID, Title เพื่อหาค่า mean
eiei = users_ratings_movies.groupby(['MovieID','Title'])['Rating'].mean()
eiei.head()
```

```python
#Ratings above 4.5
df = pd.DataFrame(eiei)
df.reset_index(level=['Title'], inplace=True)
Rating = df[df['Rating']>4.5]
Rating45 = Rating[['Title','Rating']]
(Rating45).reset_index()
```

```python
print(f'มีภาพยนตร์ทั้งหมด {len(Rating45)} เรื่องที่มีคะแนนเฉลี่ยมากกว่า 4.5')
```

### How many movies have an average rating over 4.5 among men? How about women?

```python
#Group by MovieID, Title, Gender เพื่อหาค่า mean
kk = users_ratings_movies.groupby(['MovieID','Title','Gender'])['Rating'].mean()
kk.head()
```

```python
df = pd.DataFrame(kk)
df.reset_index(level=['Gender'], inplace=True)
Rating_set = df[df['Rating']>4.5]
Ratingover45 = Rating_set[['Gender','Rating']]
(Ratingover45).reset_index()
```

```python
#คะแนนเฉลี่ยของผู้ชาย > 4.5
Ratingmen45 = Ratingover45[Ratingover45['Gender']=='M']
print(f'มีภาพยนตร์ทั้งหมด {len(Ratingmen45)} เรื่องที่มีคะแนนเฉลี่ยมากกว่า 4.5 ในหมู่ผู้ชาย')

#คะแนนเฉลี่ยของผู้หญิง > 4.5
Ratingwomen45 = Ratingover45[Ratingover45['Gender']=='F']
print(f'มีภาพยนตร์ทั้งหมด {len(Ratingwomen45)} เรื่องที่มีคะแนนเฉลี่ยมากกว่า 4.5 ในหมู่ผู้หญิง')
```
