## Stacked Barchart แสดงคะแนนรวมของนักศึกษาแต่ละคนโดยเรียงจากคะแนนรวมมากที่สุดไปน้อยที่สุด (คะแนนส่วนที่ปรึกษา ซ้อนกับ คะแนนสอบปากเปล่า)

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
figure(figsize=(10,10), dpi=80)
# create data
x = tablesort['นักศึกษา']
y1 = tablesort['คะแนนจากอาจารย์ที่ปรึกษา']
y2 = tablesort['คะแนนสอบปากเปล่า']
 
# plot bars in stack manner
plt.bar(x, y1, label ='From AJ',color='salmon')
plt.bar(x, y2, bottom=y1, label ='Oral Test',color='lightblue')
plt.xticks(tablesort.reset_index()['นักศึกษา'],rotation=90)
plt.legend()
```
