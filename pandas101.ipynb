### ต่อแบบเอาตารางมาต่อกันเลย


```python
project = pd.read_csv('/content/drive/MyDrive/BPDV65/opendata_project.csv')
project.head()
```

```python
#merge คือคำสั่งการต่อตาราง
unit.merge(project) #เพิ่มข้อมูลจากอีกตารางมาต่ออีก เอาข้อมูลจาก unit มาต่แกับ project 
#เราจะใช้คีย์คอลลัมน์ไหนมาต่อกับคอลลัมน์ไหน
```

```python
unit.merge(project , left_on='project_id', right_on='project_id') #ตารางด้านขวาต้องเป็น คีย์หลัก primary key ไม่ซ้ำในหัวตาราง
```

```python
#เราต้องการตอบคำถามที่ว่า บ้านที่แพงที่สุดในจังหวัดขอนแก่น 
this = unit.merge(project, left_on='project_id', right_on='project_id') #ตารางด้านขวาต้องเป็น คีย์หลัก primary key ไม่ซ้ำในหัวตาราง
this[this['province_name_th']=='ขอนแก่น'] #เลือกบ้านในจังหวัดขอนแก่นมาก่อน
```
