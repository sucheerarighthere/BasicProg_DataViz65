##1.เปรียบเทียบราคาเฉลี่ยของ Detached house, Condo, Twin house และ Townhome

```python
unit.head() # โชว์ข้อมูลตัวอย่างเพื่อแสดงว่าหัวตารางมีอะไรบ้าง|
```

```python
unit.groupby(['propertytype_name_en'])['price_min'].mean()[['Detached House', 'Condo', 'Twin house', 'Townhome']]
#เอาราคาเฉลี่ยออกมาก่อน เอาแค่ 'Detached House', 'Condo', 'Twin house', 'Townhome'
```

## 2.พื้นที่ (area_total_min_wa และ area_usable_min) และ จำนวนห้องนอนของทรัพย์สินที่แพงที่สุดแต่ละชนิด

```python
dm = unit.dropna(subset=['propertytype_name_en','price_min','area_total_min_wa','area_usable_min','count_room_bed']) # ลบ missing ออก 
```

```python
dm.groupby(['propertytype_name_en']).max()[['area_total_min_wa','area_usable_min','count_room_bed','price_min']]
 #Groupby ชนิดของบ้านแต่ละชนิด ใช้.max() หาค่าแพงที่สุด ใช้list เลือกเฉพาะ column พื้นที่และจำนวนห้องนอน
 #max ทีี่เอามาเป็นรายคอลลัมน์ คือมากที่สุดในแต่ละคอลลัมน์  อาจจะต้องทำต่อ เพราะ ได้ราคาmaxมาแล้ว ก็เอาราคานี้ไปหาพื้นที่ที่เราต้องการ 
```

##3.จำนวนที่จอดรถและห้องน้ำของของทรัพย์สินที่ถูกที่สุดแต่ละชนิด

```python
dm.groupby(['propertytype_name_en']).min()[['count_parking','count_room_bath','price_min']]
 #Groupby ชนิดของบ้านแต่ละชนิด ใช้.min() หาค่าต่ำที่สุด ใช้list เลือกเฉพาะ column พื้นที่และจำนวนห้องนอน
 #ยังไม่เสร็จต้องมีทำต่อเพราะเลือกมินแต่ละคอลลัมน์ออกมาอาจจะไม่ตรงกันก็ได้
```
